```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv',usecols=['Age','Pclass','SibSp','Parch','Survived'])
```

```python
df.sample(5)
```

```python
X = df.iloc[:,1:5]
y = df.iloc[:,0]
```

```python
#Fetaure Construction
X
```

```python
y
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
lr = LogisticRegression()
```

```python
np.mean(cross_val_score(lr,X,y,scoring='accuracy',cv=20))
```

```python
X['Family_Size'] = X['SibSp']+X['Parch']+1
```

```python
X.head()
```

```python
def myfucn(num):
    if num==1:
        return 0
    elif num>1 and num<4:
        return 1
    else: 
        return 2 
```

```python
myfucn(4)
```

```python
X['Family_Size'] = X['Family_Size'].apply(myfucn)
```

```python
X.head()
```

```python
X.drop(columns=['Pclass','SibSp'])
```

```python
np.mean(cross_val_score(lr,X,y,scoring='accuracy',cv=20))
```
